```python
import pandas as pd
```

```python
#pip install unidecode
from unidecode import unidecode
```

```python
df = pd.read_csv('escolas.csv')
df.head()
df['id'] = df['id'].astype(str).str.zfill(3)
df['Escolas_Postos'] = df['Escolas_Postos'].apply(lambda x: unidecode(x))
df['BAIRRO'] = df['BAIRRO'].apply(lambda x: unidecode(x))
df['ENDEREÇO '] = df['ENDEREÇO '].apply(lambda x: unidecode(x))
df['ENDEREÇO '] = df['ENDEREÇO '].str.replace('R.', 'Rua')
df['ENDEREÇO '] = df['ENDEREÇO '].str.replace('Pça.', 'Praça')
df['ENDEREÇO '] = df['ENDEREÇO '].str.replace('Praca', 'Praça')
df['ENDEREÇO '] = df['ENDEREÇO '].str.replace('Estr.', 'Estrada')
df['ENDEREÇO '] = df['ENDEREÇO '].str.replace('Av.', 'Avenida')
df['ENDEREÇO '] = df['ENDEREÇO '].str.replace('Av', 'Avenida')


# criar uma coluna chamada NUMERO_ENDERECO :
# Usamos str.extract('(\d+)') para extrair todos os números da coluna 'Endereço' ...
# ...e criar uma nova coluna 'Número' com esses números. A expressão regular (\d+) corresponde a um ou mais dígitos.
df['NUMERO_ENDERECO'] = df['ENDEREÇO '].str.extract('(\d+)')

# A coluna de numero_endereço exibe NaN quando não encontra os números. Vamos colocar S/N no lugar de NaN
# Substitua os NaN por "S/N" na coluna 'Número'
df['NUMERO_ENDERECO'] = df['NUMERO_ENDERECO'].fillna('S/N')

# TRATANDO OS NOMES DAS COLUNAS
# Remova espaços à direita nos nomes de coluna
df.columns = df.columns.str.strip()

# Renomeie todas as colunas para o formato Snake Case
df.columns = df.columns.str.upper().str.replace(' ', '_')

# Usando .upper() nas colunas para tornar todo o conteúdo em letras maiúsculas
df = df.apply(lambda x: x.str.upper())

df
```
